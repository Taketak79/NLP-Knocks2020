## 第6問

```python
def Ngram(a,n):#aは文字列,nは数字
    c = []
    for i in range(len(a)- n + 1):
        c.append(a[i:i + n])
    return c

a = input("対象の文字列は")
x = int(input("区切る文字数は"))
print(Ngram(a,x))
#空白も1文字としてカウントされる
```

## 第7問

```python
def Ngram(a,n):#aは文字列,nは数字
    c = []
    for i in range(len(a)- n + 1):
        c.append(a[i:i + n])
    return c

A = "paraparaparadise"
B = "paragraph"

X = Ngram(A,2)
Y = Ngram(B,2)

X = set(X)
Y = set(Y)

XY_union = X.union(Y)
XY_intersection = X.intersection(Y)
XY_difference = X.difference(Y)

print(XY_union)
print(XY_intersection)
print(XY_difference)
print({"se"} <= X or {"se"} <= Y)
print({"se"} <= XY_union)
```

## 第9問

```python
def cipher(text):
    c = []
    for i in text:
        if 97 <= ord(i) <= 122:
            i = chr(219 - ord(i))
        c.append(i)
    return "".join(c)

text = input("文字列を入力")
text_cipher = cipher(text)
text_cipher_cipher = cipher(text_cipher)

print(text_cipher)
print(text_cipher_cipher)
```

## 第10問

```python
import random

def Typoglycemia(text):
    c = []
    a = text.split()
    for i in a:
        if len(i) > 4:
            b = list(i[1:-1])
            random.shuffle(b)
            i = i[0] + "".join(b) + i[-1]
        c.append(i)
    d = " ".join(c)
    return d

text0 = 'I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .'
print(Typoglycemia(text0))

text = input("文字列を入力")
print(Typoglycemia(text))
```
